# Line chart

```python
import numpy as n
import matplotlib.pyplot as plt
xaxis=n.array([1,4,3,6])
yaxis=n.array([2,7,1,5])
plt.title("Line Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(xaxis,yaxis,"--",marker="*",mec="r",color="b",linewidth="2")
plt.show()
```

```python
plt.plot(n.array([1,2]))
plt.show()
```

```python
X=n.array([0,3,4,9,6,7])
Y1=n.array([2,4,5,1,12,7])
Y2=n.array([9,6,2,10,22,15])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(X,Y1,"--",marker="*",mec="r")
plt.plot(X,Y2,marker="x",mec="g")
plt.legend(["Y1","Y2"])
plt.grid()
plt.show()
```

```python
plt.scatter(X,Y1)
```

# Pie chart:

```python
y=n.array([35,50,25])
plt.pie(y)
```

```python
co=["r","b","g"]
plt.pie(y,colors=co,labels=["Wt","DSP","CD"])
plt.legend(title="subject credits")
plt.show()
```

# Bar chart:

```python
xaxis=["Maths","Science","Social","English"]
yaxis=[49,40,45,44]
plt.xlabel("Subject")
plt.title("MARKS")
plt.ylabel("MArks for subject")
plt.bar(xaxis,yaxis,color="hotpink")
plt.show()
```

# stack Bar graph:

```python
y1=n.array([1,3,5,2])
y2=n.array([2,4,3,8])
x=range(1,5)
plt.bar(x,y1)
plt.bar(x,y2,bottom=y1)
```

# Histogram chart:

```python
l=[0.1,0.5,1,1.5,2,4,5.5,6,7,8,9]
plt.xlabel("Frequency")
plt.ylabel("Numbers")
plt.title("Number-Frequency(with default bins,color)")
plt.hist(l)
```

```python
l=[0.1,0.5,1,1.5,2,4,5.5,6,7,8,9]
plt.xlabel("Frequency")
plt.ylabel("Numbers")
plt.title("Number-Frequency(with number of bins, specifiedcolor)")
plt.hist(l,color="hotpink",bins=3)
```
