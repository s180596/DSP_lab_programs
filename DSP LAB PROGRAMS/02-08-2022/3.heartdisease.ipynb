# 19

```python
l=[]
for r in df["age"]:
    if(r<15):
        l.append("young")
    elif((r>=15) & (r<35)):
        l.append("adult")
    elif((r>=35) & (r<55)):
        l.append("mid")
    else:
        l.append("old")
df["agegroup"]=l
```

```python
df
```

```python
p=pd.crosstab(df["chd"],df["agegroup"],margins=True)
p
```

```python
df.sort_values(by=["agegroup"])
```

```python
q=p.iloc[1,:]
print(pd.DataFrame(q))
```
