```python
fig=plt.figure()
x=np.arange(0,101,1)
y=np.arange(0,201,2)
plt.xticks(np.arange(0,101,20))
plt.yticks(np.arange(0,202,50))
plt.title("title")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y)
fig=plt.figure()
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
ax1.set_xlim(0.0,1.0)
ax1.set_ylim(0.0,1.0)
ax2.set_xlim(0.0,1.0)
ax2.set_ylim(0.0,1.0)
```

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(0,100)
ax1.set_ylim(0,200)
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(0,100)
ax2.set_ylim(0,200)
fig
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes( [0.2,0.5,.4,.4])
```

```python
x=np.arange(0,100)
y=x*2
z=x**2
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlim(20.0,22.0)
ax2.set_ylim(30,50)
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
fig
```

```python
x=np.arange(0,100)
y=x*2
z=x**2
plt.subplot(1,2,1)
plt.plot(x,y,"--",color="b",linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,z,color="r",linewidth=3)
plt.show()
```

```python
x=np.arange(0,100)
y=x*2
z=x**2
plt.figure(figsize=(12,2))
plt.subplot(1,2,1)
plt.plot(x,y,color="b",linewidth=3)
plt.xlabel("x")
plt.ylabel("y")
plt.subplot(1,2,2)
plt.plot(x,z,"--",color="r",linewidth=3)
plt.yticks(np.arange(0,10001,2000))
plt.xlabel("x")
plt.ylabel("z")
plt.show()
```
